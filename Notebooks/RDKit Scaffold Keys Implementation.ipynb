## find_scaffold()

Combine the above points into a callable function to use later

```python
def find_scaffold(mol):
    "RDKit mol in. RDKit mol for scaffold out"
    scaffold = Chem.MurckoDecompose(mol)
    smiles = Chem.MolToSmiles(scaffold)
    scaffold = Chem.MolFromSmiles(smiles)
    
    return scaffold
```

## Develop the functions to calculate the 32 keys

Now we have the scaffold molecule, we can start calculating the 32 scaffold keys

```python
def fully_conj_ring_bonds(scaffold, key):
    "Function for key 7 or key 8"
    ri = scaffold.GetRingInfo()
    bondRings = ri.BondRings()

    fully_conj_count = 0
    multiple_bond_not_fully_conj_count = 0
    for ring in bondRings:
        fully_conj = True
        for bondIdx in ring:
            bond = scaffold.GetBondWithIdx(bondIdx)
            if not bond.GetIsConjugated():
                fully_conj = False
                break

        if fully_conj:
            fully_conj_count += len(ring)
        else:
            multiple_bonds = 0
            for bondIdx in ring:
                bond = scaffold.GetBondWithIdx(bondIdx)
                if bond.GetBondTypeAsDouble() > 1.0:
                    multiple_bond_not_fully_conj_count += 1                
    
    if key == 7:
        return fully_conj_count
    elif key == 8:
        return multiple_bond_not_fully_conj_count
```

```python
def generate_keys(scaffold, output=False):
               
    scaffold_keys = {
    1: {
        "description": "number of ring and linker atoms",
        "scope": "RL",
        "smarts": "[!D1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!D1]")))
        },
    2: {
        "description": "number of linker atoms",
        "scope": "L",
        "smarts": "[!R;!D1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!R;!D1]")))
        },
    3: {
        "description": "number of linker bonds",
        "scope": "L",
        "smarts": "[*;!D1]~[!R;!D1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[*;!D1]~[!R;!D1]")))
        },
    4: {
        "description": "number rings",
        "scope": None, # Scope is not defined in the paper...
        "smarts": None,
        "function": Descriptors.RingCount(scaffold)
        },
    5: {
        "description": "number of spiro atoms",
        "scope": "R",
        "smarts": "[R2;x4]", # Not very happy with this as a complete definition of spirocyclic atom...
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[R2;x4]")))
        },
    6: {
        "description": "size of the largest ring",
        "scope": "R",
        "smarts": None,
        "function": max([len(i) for i in scaffold.GetRingInfo().AtomRings()])
        },
    7: {
        "description": "number of bonds in fully conjugated rings",
        "scope": "R",
        "smarts": None,
        "function": fully_conj_ring_bonds(scaffold, key=7)
        },
    8: {
        "description": "number of multiple bonds in not fully conjugated rings",
        "scope": "R",
        "smarts": None,
        "function": fully_conj_ring_bonds(scaffold, key=8)
        },
    9: {
        "description": "number of heteroatoms in rings",
        "scope": "R",
        "smarts": "[R;!#6]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[R;!#6]")))
        },
    10:{
        "description": "number of heteroatoms other than N, S, O in rings",
        "scope": "R",
        "smarts": "[R;!#6;!#7;!8;!16]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[R;!#6;!#7;!8;!16]")))
        },
    11:{
        "description": "number of S ring atoms",
        "scope": "R",
        "smarts": "[R;#16]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[R;#16]")))
        },
    12:{
        "description": "number of O ring atoms",
        "scope": "R",
        "smarts": "[R;#8]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[R;#8]")))
        },
    13:{
        "description": "number of N ring atoms",
        "scope": "R",
        "smarts": "[R;#7]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[R;#7]")))
        },
    14:{
        "description": "number of heteroatoms",
        "scope": "A",
        "smarts": None,
        "function": Descriptors.NumHeteroatoms(scaffold)
        },
    15:{
        "description": "number of heteroatoms other than N, S, O",
        "scope": "A",
        "smarts": "[!#1;!#6;!#7;!#16;!#8]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#1;!#6;!#7;!#16;!#8]")))
        },
    16:{
        "description": "number of S atoms",
        "scope": "A",
        "smarts": "[#16]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[#16]")))
        },
    17:{
        "description": "number of O atoms",
        "scope": "A",
        "smarts": "[#8]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[#8]")))
        },
    18:{
        "description": "number of N atoms",
        "scope": "A",
        "smarts": "[#7]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[#7]")))
        },
    19:{
        "description": "number of multiple linker bonds",
        "scope": "A", # Waiting to hear whether this should be 'L' instead...
        "smarts": "[!D1]!@;!-[!D1]", # For now I am going to assume that we are counting-up non-single bonds in L regions...
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!D1]!@;!-[!D1]")))
        },
    20:{
        "description": "count of two adjacent heteroatoms",
        "scope": "AO",
        "smarts": "[!#6;!#1]~[!#6;!#1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#6;!#1]~[!#6;!#1]")))
        },
    21:{
        "description": "count of 3 adjacent heteroatoms",
        "scope": "AO",
        "smarts": "[!#6;!#1]~[!#6;!#1]~[!#6;!#1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#6;!#1]~[!#6;!#1]~[!#6;!#1]")))
        },
    22:{
        "description": "count of 2 heteroatoms separated by a single carbon",
        "scope": "AO",
        "smarts": "[!#6;!#1]~[#6]~[!#6;!#1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#6;!#1]~[#6]~[!#6;!#1]")))
        },
    23:{
        "description": "count of 2 heteroatoms separated by 2 carbons",
        "scope": "AO",
        "smarts": "[!#6;!#1]~[#6]~[!#6;!#1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#6;!#1]~[#6]~[#6]~[!#6;!#1]")))
        },
    24:{
        "description": "number of double bonds with at least one heteroatom",
        "scope": "AO",
        "smarts": "[!#6;!#1]=*",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#6;!#1]=*")))
        },
    25:{
        "description": "number of heteroatoms adjacent to a double (nonaromatic) bond",
        "scope": "AO",
        "smarts": "[!#6;!#1]-*=;!:*",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#6;!#1]-*=;!:*")))
        },
    26:{
        "description": "count of pairs of conjugated double (nonaromatic) bonds",
        "scope": "AO",
        "smarts": "*=;!:*-*=;!:*",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("*=;!:*-*=;!:*")))
        },
    27:{
        "description": "count of pairs of adjacent branched atoms",
        "scope": "AO",
        "smarts": "[!D1;!D2]~[!D1;!D2]", # I've interpreted 'branched' atom to be any atom that has degree > 2
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!D1;!D2]~[!D1;!D2]")))
        },
    28:{
        "description": "count of branched atoms separated by a single nonbranched atom",
        "scope": "AO",
        "smarts": "[!D1;!D2]~[D2]~[!D1;!D2]", # I've interpreted 'branched' atom to be any atom that has degree > 2
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!D1;!D2]~[D2]~[!D1;!D2]")))
        },
    29:{
        "description": "count 3 adjacent branched atoms",
        "scope": "AO",
        "smarts": "[!D1;!D2]~[!D1;!D2]~[!D1;!D2]", # I've interpreted 'branched' atom to be any atom that has degree > 2
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!D1;!D2]~[!D1;!D2]~[!D1;!D2]")))
        },
    30:{
        "description": "count of branched atoms separated by any 2 atoms",
        "scope": "AO",
        "smarts": "[!D1;!D2]~*~*~[!D1;!D2]", # I've interpreted 'branched' atom to be any atom that has degree > 2
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!D1;!D2]~*~*~[!D1;!D2]")))
        },
    31:{
        "description": "number of exocyclic and exolinker atoms",
        "scope": "!R, !L",
        "smarts": "[D1]",
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[D1]")))
        },
    32:{
        "description": "number of heteroatoms with more than 2 bonds",
        "scope": "R", # The scope says 'R', but the name suggests otherwise...
        "smarts": "[!#6;!#1;!D1;!D2]", # For now, I am going to interpret as any scope (R or L) with degree > 2
        "function": len(scaffold.GetSubstructMatches(Chem.MolFromSmarts("[!#6;!#1;!D1;!D2]")))
        }
    }
    
    od = collections.OrderedDict(sorted(scaffold_keys.items()))
    keys = []
    for k in od:
        keys.append(od[k]['function'])
        if output:
            print "Rule #%02d:  %s (%s)" % (k, od[k]['function'], od[k]['description'])
    
    return keys
```

## Download some test molecules from ChEMBL

Now lets test this out with the first 1000 (well, if they exist, and if RDKit is happy) structures in ChEMBL

```python
import requests
test_mols = {}

for n in range(1, 1001):
    address = "https://www.ebi.ac.uk/chemblws/compounds/CHEMBL%s.json" % (n,)
    try:
        r = requests.get(address)
        smiles = r.json()['compound']['smiles']
        chemblId = r.json()['compound']['chemblId']
        mol = Chem.MolFromSmiles(smiles)
    except:
        print "x",
        pass
    else:
        test_mols[chemblId] = mol
        print ".",

print "\n"
print "%s structures retrieved from ChEMBL." % (len(test_mols),)

    
```

This will print the ChEMBL IDs against a string representation of the scaffold keys.  We can then use this string representation as a quick way of sorting the scaffolds

```python
import collections
results = collections.OrderedDict()

for k, v in test_mols.iteritems():
    
    try:
        scaffold = find_scaffold(v)
        scaffold_keys_list = generate_keys(scaffold)
        scaffold_keys_string = "".join(["%03d" % (val,) for val in scaffold_keys_list])
    except:
        results[k] = None
    else:
        results[k] = {'scaffold': scaffold,
                      'scaffold_keys_list': scaffold_keys_list,
                      'scaffold_keys_string': scaffold_keys_string}

for k, v in results.iteritems():
    if v:
        print "%s:  %s" % (k, v['scaffold_keys_string'])
```

```python
results_list = {}
for k, v in results.iteritems():
    if not v:
        continue
    if not results_list.has_key(v['scaffold_keys_string']):
        results_list[v['scaffold_keys_string']] = []
    results_list[v['scaffold_keys_string']].append((k, v['scaffold']))
    
results_list_sorted = collections.OrderedDict(sorted(results_list.items()))

print "%s scaffolds identified" % (len(results_list_sorted),)
```

```python
output = []
legends = []
for k, v in results_list_sorted.iteritems():
    if len(v) == 1:
        plural = ""
    else:
        plural = "s"
    legends.append("%s example%s" % (len(v), plural))
    output.append(v[0][1]) # first copy of the scaffold

Draw.MolsToGridImage(output, legends=legends)
```
